# Simple trading strategy using Moving Average

## Objective
The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price. The average is taken over a specific period of time, like 10 days, 20 minutes, 30 weeks or any time period the trader chooses.

MA of short period is more closely associated with recent changes of stock price, then called as  **fast signal**. MA over long period reflects the price change over long-term history then called as **slow signal**.

Some traders believes if fast signal > slow signal the stock price will goes up in next several days. Otherwise, the price will decrease. Using this strategy we will try to make some profit based on MA signal.

## Case Study
Stock Code : BBCA (IDX Market)

Training data period: Jan 1, 2019 - Dec 31,2019

Strategy : If fast signal > slow signal, we will buy and hold one share of stock. 


```python
# to handle dataset
import pandas as pd
import datetime as dt

#to visualize
import matplotlib.pyplot as plt
%matplotlib inline

#to crawl stock data
import pandas_datareader.data as web
```

### 1. Munging the stock data and add two columns [MA(20), MA[50]]

```python
# load data BBCA since 2019 until first 2020
start = dt.datetime(2019,1,1)
#end = dt.datetime.now() #exclude anomalies due to covid-19
end = dt.datetime(2020,1,1)

bbca = web.DataReader("BBCA.JK", 'yahoo', start,end)
bbca['ma7'] = bbca['Close'].rolling(3).mean()
bbca['ma30'] = bbca['Close'].rolling(21).mean()
bbca = bbca.dropna()

# plot the moving average
bbca['Close'].loc['2018-01-01':'2020-01-01'].plot(label='Close')
bbca['ma7'].plot(label='ma7') #magic number for short and long period
bbca['ma30'].plot(label='ma30') 

plt.legend()
plt.show()
```

### 2. Add 'shares' column to make decisions base on the startegy

if MA15 > MA30, detones as 1 which means we buy one share stock. Otherwise, denote as 0 as do nothing

```python
bbca['Shares'] = [1 if bbca.loc[ei, 'ma7'] > bbca.loc[ei,'ma30'] else 0 for ei in bbca.index]
```

If **Shares =1** the **Profit** is calculated as the close price of tomorrow minus the close price of today. Otherwise the profit is 0

```python
# plot a graph to show the profit/loss
bbca['CloseLead1'] = bbca['Close'].shift(-1)
bbca['Profit'] = [bbca.loc[ei,'CloseLead1'] - bbca.loc[ei,'Close'] 
                  if bbca.loc[ei,'Shares']==1 else 0 for ei in bbca.index]
bbca['Profit'].plot()
plt.axhline(y=0, color='red')
```

```python
#calculate the accumulated wealth over the period using cumsum()
bbca['CumProfit'] = bbca['Profit'].cumsum()

#plot accumulated of profit over the period
bbca['CumProfit'].plot()
plt.title('Total money you gain is {}'.format(bbca.loc[bbca.index[-2], 'CumProfit']))
```

```python
print('Total Asset: {}'.format(bbca['CloseLead1'].sum()))
print('Total Capital Gain (if you sell your stocks):{}')
```
